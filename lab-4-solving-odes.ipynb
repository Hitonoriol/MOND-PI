### 1. Write a program that computes the solution `y(t)` of this ODE using the odeint algorithm: `dy / dt = −exp(−t) * (10 * sin(10 * t) + cos(10 * t))` for `t = 0..10`. The initial value is `y(0) = 1`.

```python
from scipy.integrate import odeint
from numpy import exp, sin, cos
import matplotlib.pyplot as plt
import numpy as np
import pylab

f = lambda y, t: -exp(-t) * (10 * sin(10 * t) + cos(10 * t))

y0 = 1 # Initial value
a, b = 0, 10 # Integration range bounds
step = 0.01

t = np.arange(a, b + step, step) # Inclusive integration range [a; b]
y = odeint(f, y0, t) # Solve the ODE dy / dt = f(y, t)

print(f"Computed points of y(t) for each t = {a}..{b} with step = {step}:")
print(y)
```

### 2. You should display the solution graphically at points `t = 0`, `t = 0.01`, `t = 0.02`, ..., `t = 9.99`, `t = 10`.

```python
print(f"Values of `t` to plot y(t) for:\n{t}")
fig, ax = plt.subplots()
ax.plot(t, y, label = "y(t)")
ax.set(xlabel = 't', ylabel = "y(t)", title = "ODE Solution")
ax.grid()
fig.show()
```
